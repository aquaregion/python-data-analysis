```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/pydata/pydata-book/master/ch02/names/yob1880.txt",names=["name","gender","count"])
data['year'] = '1880'
```

```python
for i in range(1881,2011):
    tmp = pd.read_csv("https://raw.githubusercontent.com/pydata/pydata-book/master/ch02/names/yob%d.txt"%i,names=["name","gender","count"])
    tmp['year'] = str(i)
    data = pd.concat([data,tmp])
```

```python
data['ratio'] = data.groupby(['year','gender']).transform(lambda x:x/x.sum())
```

```python
most_common_names = data.groupby(['year','gender']).aggregate({'ratio':'max','name':'first'})
```

```python
most_common_names.reset_index(inplace=True)
```

```python
fig,ax = plt.subplots(figsize=(8,6))

ax.plot(most_common_names[most_common_names['gender']=='M']['year'],most_common_names[most_common_names['gender']=='M']['ratio'])
for row in most_common_names[most_common_names['gender']=='M'][::5].iterrows():
    ax.text(int(row[1]['year']), row[1]['ratio'],row[1]['name'])
```

```python
fig,ax = plt.subplots(figsize=(8,6))

ax.plot(most_common_names[most_common_names['gender']=='F']['year'],most_common_names[most_common_names['gender']=='F']['ratio'])
for row in most_common_names[most_common_names['gender']=='F'][::5].iterrows():
    ax.text(int(row[1]['year']), row[1]['ratio'],row[1]['name'])
```

```python
most_common_names.groupby(['gender'])['name'].value_counts().plot(kind='barh',by=['gender'])
```

```python
most_common_names.pivot_table(columns=['year','gender','name'],values=['ratio'],)
```
